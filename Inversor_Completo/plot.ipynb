```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

# Configurações de plot:
plt.style.use([
    'grid',
    'retro'
])
#plt.rcParams['lines.linewidth'] = 3
#plt.rcParams['font.size'] = 20
plt.rcParams['figure.figsize'] = (12, 5)
```

```python
corrente_motor = np.loadtxt('corrente_motor.txt')    #Time	Ia	Ib	Ic
fase_trifasico = np.loadtxt('fase_trifasico.txt')    #Time	Fase
freq = np.loadtxt('freq.txt')    #Time	Freq
ref_vel = np.loadtxt('ref_vel.txt')    #Time	Referencia	Velocidade
tensao_fase_fase = np.loadtxt('tensao_fase-fase.txt')    #Time	Vab	Vfase1-vfase2
triangular_vabc = np.loadtxt('triangular_vabc_pwm.txt')    #Time	Triangular	Va	Vb	Vc
vabc_trifasico = np.loadtxt('vabc_trifasico.txt')    #Time	Va	Vb	Vc


tf = 4
```

```python
plt.figure(1)
plt.plot((corrente_motor[:, 0]), corrente_motor[:, 1], label='$I_a$')
plt.plot((corrente_motor[:, 0]), corrente_motor[:, 2], label='$I_b$')
plt.plot((corrente_motor[:, 0]), corrente_motor[:, 3], label='$I_c$')
plt.ylabel('Corrente [A]')
plt.xlabel('Tempo [s]')
plt.xlim(0, tf)
plt.legend()
plt.savefig('curves/corrente_motor.png', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(2)
plt.plot((fase_trifasico[:, 0]), fase_trifasico[:, 1], color='C4')
plt.ylabel('Fase [°]')
plt.xlabel('Tempo [s]')
plt.xlim(0, 1)
#plt.legend()
plt.savefig('curves/fase_trifasico.png', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(3)
plt.subplots(2, 1, sharex=True)
plt.subplot(2, 1, 1)
plt.plot((vabc_trifasico[:, 0]), vabc_trifasico[:, 1], label='$V_a$')
plt.plot((vabc_trifasico[:, 0]), vabc_trifasico[:, 2], label='$V_b$')
plt.plot((vabc_trifasico[:, 0]), vabc_trifasico[:, 3], label='$V_c$')
plt.ylabel('Tensão [V]')
plt.legend()
plt.subplot(2, 1, 2)
plt.plot((freq[:, 0]), freq[:, 1], color='C5')
plt.ylabel('Frequência [Hz]')
plt.xlabel('Tempo [s]')
plt.xlim(0, 2.5)
plt.savefig('curves/freqxtensao.png', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(4)
plt.plot((ref_vel[:, 0]), ref_vel[:, 1], label='$V_{ref}$')
plt.plot((ref_vel[:, 0]), ref_vel[:, 2], label='$V$')
plt.ylabel('Velocidade [RPM]')
plt.xlabel('Tempo [s]')
plt.xlim(0, tf)
plt.legend(loc='center right')
plt.savefig('curves/ref_vel.png', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(5)
plt.plot((tensao_fase_fase[:, 0]), tensao_fase_fase[:, 1], label='PWM')
plt.plot(((tensao_fase_fase[:, 0])-0.007), tensao_fase_fase[:, 2], label='LPF')
plt.ylabel('Tensão [V]')
plt.xlabel('Tempo [s]')
plt.xlim(0, 0.08)
plt.legend()
plt.savefig('curves/tensao_fase_fase.png', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(6)
plt.plot((triangular_vabc[:, 0]), triangular_vabc[:, 1], label='$Triangular$', color='C4')
plt.plot((triangular_vabc[:, 0]), triangular_vabc[:, 2], label='$V_a$')
plt.plot((triangular_vabc[:, 0]), triangular_vabc[:, 3], label='$V_b$')
plt.plot((triangular_vabc[:, 0]), triangular_vabc[:, 4], label='$V_c$')
plt.ylabel('Tensão [V]')
plt.xlabel('Tempo [s]')
plt.xlim(0, 2)
plt.legend()
plt.savefig('curves/triangular_vabc.png', dpi=600, transparent=True, bbox_inches='tight')
```
