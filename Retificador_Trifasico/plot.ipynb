```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

# Configurações de plot:
plt.style.use([
    'grid',
    'retro'
])
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['font.size'] = 16
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
vlinha = np.loadtxt('data/vlinha.txt')   #Time    V_a     V_b     V_c
vfase = np.loadtxt('data/vfase.txt')   #Time	    V_ab	V_bc	V_ca
```

```python
plt.figure(1)
plt.subplots(2, 1, sharex=True)
plt.subplot(2, 1, 1)
plt.plot(vlinha[:, 0]*1e3, vlinha[:, 1], label='$V_a$')
plt.plot(vlinha[:, 0]*1e3, vlinha[:, 2], label='$V_b$')
plt.plot(vlinha[:, 0]*1e3, vlinha[:, 3], label='$V_c$')
plt.ylabel('Tensão [V]')
plt.legend()
plt.subplot(2, 1, 2)
plt.plot(vfase[:, 0]*1e3, vfase[:, 1], color='C3', label='$V_{ab}$')
plt.plot(vfase[:, 0]*1e3, vfase[:, 2], color='C4', label='$V_{bc}$')
plt.plot(vfase[:, 0]*1e3, vfase[:, 3], color='C5', label='$V_{ca}$')
plt.ylabel('Tensão [V]')
plt.xlabel('Tempo [ms]')
plt.xlim(0, (2/60)*1e3)
plt.legend()
plt.show()
```

```python
plt.figure(2)
plt.plot(vfase[:, 0]*1e3, vfase[:, 1], label='$V_{ab}$')
plt.plot(vfase[:, 0]*1e3, vfase[:, 2], label='$V_{bc}$')
plt.plot(vfase[:, 0]*1e3, vfase[:, 3], label='$V_{ca}$')
plt.ylabel('Tensão [V]')
plt.xlabel('Tempo [ms]')
plt.xlim(0, (2/60)*1e3)
plt.legend()
plt.show()
```
