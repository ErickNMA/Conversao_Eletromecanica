```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

# Configurações de plot:
plt.style.use([
    'grid',
    'retro'
])
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['font.size'] = 16
plt.rcParams['figure.figsize'] = (15, 6)
```

```python
vlinha = np.loadtxt('data/vlinha.txt')          #Time   V_a         V_b             V_c
vfase = np.loadtxt('data/vfase.txt')            #Time	V_ab	    V_bc	        V_ca
fases = np.loadtxt('data/fases.txt')            #Time	Fase_ab	    Fase_ac	        Fase_ba	    Fase_bc	    Fase_ca	    Fase_cb
disparos = np.loadtxt('data/disparos.txt')      #Time	Disparo_ab  Disparo_ac      Disparo_ba  Disparo_bc  Disparo_ca	Disparo_cb
correntes = np.loadtxt('data/correntes.txt')    #Time	Io	        Io_filtrado	    Io_ref
tensoes = np.loadtxt('data/tensoes.txt')        #Time	V_out	    V_out_filtrado	Vt_ref
tzoom = np.loadtxt('data/tzoom.txt')          #Time	V_out
```

```python
plt.figure(1)
plt.subplots(2, 1, sharex=True)
plt.subplots_adjust(hspace=0.05)
plt.subplot(2, 1, 1)
plt.plot(vlinha[:, 0]*1e3, vlinha[:, 1], label='$V_a$')
plt.plot(vlinha[:, 0]*1e3, vlinha[:, 2], label='$V_b$')
plt.plot(vlinha[:, 0]*1e3, vlinha[:, 3], label='$V_c$')
plt.ylabel('Tensão [V]')
plt.legend()
plt.subplot(2, 1, 2)
plt.plot(vfase[:, 0]*1e3, vfase[:, 1], color='C3', label='$V_{ab}$')
plt.plot(vfase[:, 0]*1e3, vfase[:, 2], color='C4', label='$V_{bc}$')
plt.plot(vfase[:, 0]*1e3, vfase[:, 3], color='C5', label='$V_{ca}$')
plt.ylabel('Tensão [V]')
plt.xlabel('Tempo [ms]')
plt.xlim(0, (2/60)*1e3)
plt.legend()
plt.savefig('curves/tin.eps', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(2)
plt.subplots(2, 1, sharex=True)
plt.subplots_adjust(hspace=0.05)
plt.subplot(2, 1, 1)
plt.plot(fases[:, 0]*1e3, fases[:, 1], label='$\phi_{ab}$')
plt.plot(fases[:, 0]*1e3, fases[:, 2], label='$\phi_{ac}$')
plt.plot(fases[:, 0]*1e3, fases[:, 3], label='$\phi_{ba}$')
plt.plot(fases[:, 0]*1e3, fases[:, 4], label='$\phi_{bc}$')
plt.plot(fases[:, 0]*1e3, fases[:, 5], label='$\phi_{ca}$')
plt.plot(fases[:, 0]*1e3, fases[:, 6], label='$\phi_{cb}$')
plt.plot((fases[:, 0][0]*1e3, fases[:, 0][-1]*1e3), (60, 60), '--r', label='Threshold')
plt.ylabel('Fase [°]')
plt.legend(loc='lower right')
plt.subplot(2, 1, 2)
plt.plot(disparos[:, 0]*1e3, disparos[:, 1], label='$D_{ab}$')
plt.plot(disparos[:, 0]*1e3, disparos[:, 2], label='$D_{ac}$')
plt.plot(disparos[:, 0]*1e3, disparos[:, 3], label='$D_{ba}$')
plt.plot(disparos[:, 0]*1e3, disparos[:, 4], label='$D_{bc}$')
plt.plot(disparos[:, 0]*1e3, disparos[:, 5], label='$D_{ca}$')
plt.plot(disparos[:, 0]*1e3, disparos[:, 6], label='$D_{cb}$')
plt.ylabel('Sinal de disparo')
plt.xlabel('Tempo [ms]')
plt.xlim(155, 190)
plt.legend(loc='lower right')
plt.savefig('curves/fasesxdisparo.eps', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(3)
plt.plot(tensoes[:, 0]*1e3, tensoes[:, 1], label='$V_o(t)$')
plt.plot(tensoes[:, 0]*1e3, tensoes[:, 3], label='$V_{ref}(t)$')
plt.plot(tensoes[:, 0]*1e3, tensoes[:, 2], label='$LPF$')
plt.ylabel('Tensão [V]')
plt.xlabel('Tempo [ms]')
plt.legend()
plt.savefig('curves/tensoes.eps', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(4)
plt.plot(tzoom[:, 0]*1e3, tzoom[:, 1], label='$V_o(t)$')
plt.ylabel('Tensão [V]')
plt.xlabel('Tempo [ms]')
plt.xlim(100, 111)
plt.ylim(180, 310)
plt.legend()
plt.savefig('curves/tzoom.eps', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(5)
plt.plot(correntes[:, 0]*1e3, correntes[:, 1], label='$i_o(t)$')
plt.plot(correntes[:, 0]*1e3, correntes[:, 3], label='$i_{ref}(t)$')
plt.plot(correntes[:, 0]*1e3, correntes[:, 2], label='$LPF$')
plt.ylabel('Corrente [A]')
plt.xlabel('Tempo [ms]')
plt.legend()
plt.savefig('curves/correntes.eps', dpi=600, transparent=True, bbox_inches='tight')
```
