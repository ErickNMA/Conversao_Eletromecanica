<h1 align='center'> Partida em Rampa <h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

# Configurações de plot:
plt.style.use([
    'grid',
    'retro'
])
#plt.rcParams['lines.linewidth'] = 3
#plt.rcParams['font.size'] = 20
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
partida_delta = np.loadtxt('data/partida_rampa.txt')    #Time, freq, IA, IB, IC, torque, Va, Vb, Vc, vel

tf = 5
```

```python
plt.figure(1)
plt.plot((partida_delta[:, 0]), partida_delta[:, 6], label='$V_a$')
plt.plot((partida_delta[:, 0]), partida_delta[:, 7], label='$V_b$')
plt.plot((partida_delta[:, 0]), partida_delta[:, 8], label='$V_c$')
#plt.title('')
plt.ylabel('Tensão [V]')
plt.xlabel('Tempo [s]')
plt.xlim(0, tf)
plt.legend()
plt.savefig('curves/tensao_rampa.png', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(2)
plt.plot((partida_delta[:, 0]), partida_delta[:, 1], color='C2')
#plt.title('')
plt.ylabel('Frequência [Hz]')
plt.xlabel('Tempo [s]')
plt.xlim(0, tf)
#plt.legend()
plt.savefig('curves/freq_rampa.png', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(3)
plt.plot((partida_delta[:, 0]), partida_delta[:, 2], label='$I_a$')
plt.plot((partida_delta[:, 0]), partida_delta[:, 3], label='$I_b$')
plt.plot((partida_delta[:, 0]), partida_delta[:, 4], label='$I_c$')
#plt.title('')
plt.ylabel('Corrente [A]')
plt.xlabel('Tempo [s]')
plt.xlim(0, tf)
plt.legend()
plt.savefig('curves/corrente_rampa.png', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(4)
plt.plot((partida_delta[:, 0]), partida_delta[:, 9], color='C4')
plt.plot()
#plt.title('')
plt.ylabel('Velocidade [RPM]')
plt.xlabel('Tempo [s]')
plt.xlim(0, tf)
#plt.legend()
plt.savefig('curves/vel_rampa.png', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(5)
plt.plot((partida_delta[:, 0]), partida_delta[:, 5], color='C5')
#plt.title('')
plt.ylabel('Torque $[N \cdot m]$')
plt.xlabel('Tempo [s]')
plt.xlim(0, tf)
#plt.legend()
plt.savefig('curves/torque_rampa.png', dpi=600, transparent=True, bbox_inches='tight')
```

```python
plt.figure(6)
plt.plot(partida_delta[:, 9], partida_delta[:, 5], color='C7')
#plt.title('')
plt.ylabel('Torque $[N \cdot m]$')
plt.xlabel('Velocidade [RPM]')
#plt.legend()
plt.savefig('curves/tqxtp_rampa.png', dpi=600, transparent=True, bbox_inches='tight')
```
